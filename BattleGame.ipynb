```python
# Imports
import random as rand
from enum import Enum

# Random stuff
randNum = rand.randint(0,6)
flag = True


#Classes
class Enemy:
    
    def __init__(self, name: str, health: int, damage: int, level: int):
        self.name = name
        self.health = health
        self.damage = damage
        self.level = level
    
    def attack(self, target) -> None: 
        target.hp -= self.damage
        target.hp = max(target.hp, 0)    
        
        
class Character:
    
    def __init__(self, name: str, health: int, damage: int, level: int):
        self.name = name
        self.health = health
        self.damage = damage
        self.level = level
        
    def attack(self, target) -> None: 
        target.hp -= self.damage
        target.hp = max(target.hp, 0)

# Enemies
Bird = Enemy(name="Bird", health=100, damage=3, level=1)
Crocodile = Enemy(name="Crocodile", health=125, damage=5, level=1)

enemies = [Bird, Crocodile]

# Functions
def get_encounter():
    Opponent = rand.choice(enemies)
    print("You have encountered a", Opponent.name)

def get_continue():
    print("Continue?")
    print("Yes [1] No [2]")
    choice = int(input())
    if choice == 1:
        return
    elif choice == 2:
        print("Exiting")
        
        
# Main program
print("-"*30)
print("You walk into some grass, every step has a 1 in 5 chance to give you an encounter")
print("-"*30)

while flag:
    print("Move forward?")
    get_continue()
    Forward = int(input())
    if Forward == 1:
        print("Moving...")
        randNum = rand.randint(0,6)
        if randNum == 5:
            print("You have gotten an encounter!")
            get_encounter()
            flag = False
        else:
            print("You didn't get an encounter...")
            print("Move forward again?")
            get_continue()

    elif Forward == 2:
        print("Exiting")
        flag = False
```
