```python
# Imports
import random as rand
from enum import Enum
import sys


# Random stuff
randNum = rand.randint(0,6)
anger = 0
party = 0


# Items
items = 0
HealthPotion = {
    "name": "Health Potion",
    "value": 20
}
DamagePotion = {
    "name": "Damage Potion",
    "value": 20 
}
ItemList = [HealthPotion, DamagePotion]

#Classes
class Enemy:
    
    def __init__(self, name: str, health: int, damage: int, level: int):
        self.name = name
        self.health = health
        self.damage = damage
        self.level = level
    
    def attack(self, target) -> None: 
        target.health -= self.damage
        target.health = max(target.health, 0)    
        
        
class Character:
    
    def __init__(self, name: str, health: int, damage: int, level: int):
        self.name = name
        self.health = health
        self.damage = damage
        self.level = level
        
    def attack(self, target) -> None: 
        target.health -= self.damage
        target.health = max(target.health, 0)

# Enemies
Bird = Enemy(name="Bird", health=100, damage=10, level=rand.randint(2,5))
Crocodile = Enemy(name="Crocodile", health=125, damage=15, level=rand.randint(3,7))

enemies = [Bird, Crocodile]

# Characters

Flastel = Character(name="Flastel", health=125, damage=5, level=5)
Orgal = Character(name="Orgal", health=100, damage=10, level=5)
Chudzie = Character(name="Chudzi", health=75, damage=15, level=5)

characters = [Flastel, Orgal, Chudzie]
# Functions
def get_item():
    global items
    global randNum
    
    itemChance = rand.randint(1,5)
    
    if itemChance == 2:
        item = rand.choice(ItemList)
        print("-"*15)
        print("You have picked up a...")
        print(item["name"])
        print("-"*15)
        
        items += 1
    


def get_encounter():
    Opponent = rand.choice(enemies)
    print("-"*30)    
    print("You have encountered a", Opponent.name)
    print(
        "\nName:", Opponent.name,
        "\nHealth:", Opponent.health,
        "\nDamage:", Opponent.damage,
        "\nLevel:", Opponent.level                
          )
    
    print("You choose", PartyMembers.name)
    
    print("What would you like to do?")
    print(
        "\n[1] Attack"
        "\n[2] Switch"
        "\n[3] Item"
        "\n[4] Run"
    )
    FightChoice = int(input())
    if FightChoice == 1:
        print("You choose to attack!")
    elif FightChoice == 2:
        if party > 1:
            print("You choose to switch")
        else:
            print("You do not have enough party members to switch.")
            return FightChoice            
    elif FightChoice == 3:
        print("You choose to use an item")
        if items >= 1:
            print("What would you like to use?")
            print(ItemList)
    elif FightChoice == 4:
        print("You choose to run! ()")   
        
        
def get_continue():
    print("Yes [1] No [2]")
    choice = int(input())
    if choice == 1:
        print("Continuing...")
        print("-"*30)
        return 
    elif choice == 2:
        print("Exiting")
        
        
# Main program
flag = True

while flag:
    
    print("Choose the character you want to use:")
    print("\nFlastel [1] - more health, less damage",
        "\nOrgal [2] - average health, average damage",
        "\nChudzie [3] - less health, more damage")
    
    CharacterChoice = int(input())
    if CharacterChoice == 1:
        print("You have picked Flastel!")
        party = party + 1
        PartyMembers = Flastel
        flag = False
    elif CharacterChoice == 2:
        print("You have picked Orgal!")
        party = party + 1
        PartyMembers = Orgal
        flag = False
    elif CharacterChoice == 3:
        print("You have picked Chudzie!")
        party = party + 1
        PartyMembers = Chudzie
        flag = False
    else:
        print("That wasn't an option.")
        anger = anger + 1
        print("Try again")
        flag = True

    if anger >= 3:
        print("Get out!")
        flag = False
        break

if anger < 3:
    
    print("-"*30)
    print("You walk into some grass, every step has a 1 in 5 chance to give you an encounter")
    print("-"*30)   
    
    flag = True
    
    while flag:  
        
        print("Move forward?")
        Forward = int(input())
        if Forward == 1:
            print("Moving...")
            get_item()
            
            randNum = rand.randint(0,6)
            
            if randNum == 5:
                print("You have gotten an encounter!")
                get_encounter()
                flag = False
            else:
                print("You didn't get an encounter...")
                print("Move forward again?")
                get_continue()

        elif Forward == 2:
            print("Exiting")
            flag = False
        else:
            print("That wasn't an option")
            flag = True    

```
