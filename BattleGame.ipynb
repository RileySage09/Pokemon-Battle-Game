```python
# Imports
import random as rand

#
randNum = rand.randint(0,6)
flag = True

#Classes
class Enemy:
    
    def __init__(self, name: str, health: int, damage: int, level: int):
        self.name = name
        self.health = health
        self.damage = damage
        self.level = level
        
        
        
class Character:
    
    def __init__(self, name: str, health: int, damage: int, level: int):
        self.name = name
        self.health = health
        self.damage = damage
        self.level = level
        




# Functions
def get_encounter():
    print("You have encountered a bird!")
    print("Bird - 20/20")
    

def get_continue():
    print("Yes [1] No [2]")
    choice = int(input())
    if choice == 1:
        return
    elif choice == 2:
        print("Exiting")
# Main program
print("-"*30)
print("You walk into some grass, every step has a 1 in 5 chance to give you an encounter")
print("-"*30)
while flag:
    print("Move forward?")
    get_continue()
    Forward = int(input())
    if Forward == 1:
        print("Moving...")
        randNum = rand.randint(0,6)
        if randNum == 5:
            print("You have gotten an encounter!")
            get_encounter()
        elif not randNum == 5:
            print("You didn't get an encounter...")
            print("Move forward again?")
            get_continue()

    elif Forward == 2:
        print("Exiting")
        flag = False
```
